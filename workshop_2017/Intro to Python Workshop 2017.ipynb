## Populating Arrays

You don't always have to put in the values of your array manually, especially if, for example, you want a function to sample numbers evenly along some axis. There are a couple of ways to do this, but the best ones are the "x space" functions. Linspace will give you linearly spaced values given a range, and logspace will give you log-10 spaced values given a range. The arguments are 1) your beginning number, 2) your end number, and 3) how many data points you want - these functions are *inclusive*, so the last value of your array is your end number. Note that in logspace you give a *log range* so if you want to sample from 1 to 100, your range should be 0, 2.

```python
lin_array = np.linspace(0, 10, 11)
log_array = np.logspace(0, 2, 10)
print(lin_array, log_array)
```

You can use arange as well, but this is *exclusive* - it will sample numbers linearly beginning with your initial number and end one step size before your end number. Confusingly, the third argument here is your step size, not your number of data points.

```python
my_arange = np.arange(1, 10, 1)
print(my_arange)
```

Another way to create an array quickly is through np.zeros. Zeros populates an array with, well, zeros. Sounds useless but it is a great way to create an empty n-dimensional array whose values can later be replaced with outputs from some function you build. Spoiler alert...

# Control Structure

Let's move on to something a little more interesting!

Two fundamental parts of any coding language are the conditional if statement and the for loop.

The if statement is exactly what it sounds like: if something is true, then code should respond in x way. If something is not true, then code should do something else.

The for loop is a function that loops along an index for everything in some list or array.

## If statement

The simplest if statement is just a boolean (true or false).

'==' is the symbol for equivalence, producing the boolean response. You need '=' if you're writing a definition.

'!=' is the symbol for non-equivalence.

Let's try a very simple one: if 1 definitely equals 1, print yes.

```python
if 1 == 1:
    print('yes!')
```

```python
if 1 != 1:
    print('no!')

if 1 == 0:
    print('no!')
```

Note how this did nothing, since it is NOT true that 1 does not equal 1, and it is NOT true that 1 is equivalent to 0.

But we can build on this statement with "else":

```python
if 1 != 1:
    print('yes!')
else:
    print('1 always equals 1, dummy')
    
#######
a = 2

if a == 1:
    print("a=1")
else:
    print("a=2")
```

If statements follow basic boolean logic. Be careful with "or" statements -- be explicit as possible. In boolean logic, OR requires exactly one of the statements to be true.

```python
b = 10.

if b == 10 or b == 2.:
    print('b=10')
else:
    print('the or statement didnt work')
```

In addition to equivalence, Python knows less than/less than or equal to (<, <=) and greater than/greater than or equal to (>, >=).

```python
if lin_array[0] < 1:
    print('yes')
```

If statements don't do very much on their own, though, and it's more useful to embed them in a for loop.


## For Loops

The syntax for a for loop is "for x in y:".

It's fine to write "for x in y" when going over a list, but arrays sometimes have trouble with this, for reasons unclear to me. To loop over an array, it's best to use this syntax:

    for i in range(0, len(y)):
    
This says "loop over the index range of the first index in the array until the end of the array."

```python
for characters in beemovie:
    print(characters)
    
for i in lin_array:
    print(i)
    
#It worked here but don't expect it to work for more complicated functions.
```

```python
#Note that since we changed the definition of the loop, 'i' isn't the array element, 
#it's the index number:

for i in range(0, len(lin_array_plus2)):
    print(i, lin_array_plus2[i])
    
for i in range(4, 6):
    print(i, lin_array_plus2[i])
```

As promised, you can embed if statements within a for loop using a tab on the next line. We can write something that searches the array for certain values. We're adding the "elif" statement here, which is short for "else if:". This means: if the first statement isn't true but the second one is true, do X. "else" means "if anything else is true."


```python
for i in range(0, len(lin_array)):
    if lin_array[i] + 2. < 5.: #You can put arithmetic inside the statement
        print(lin_array[i] + 2, "< 5")
    elif lin_array[i] + 2. == 5.: 
        print(lin_array[i] + 2, "= 5")
    elif lin_array[i] + 2. > 5.:
        print(lin_array[i], "+ 2> 5")
```

# User-defined Functions

The most important tool you will use when programming is defining your own function. This will be the basis of any code you write: a series of functions that accomplish some task.

Functions can be defined for literally anything: mathematical expressions, list/array sorting, MCMC codes, statistics, etc. Let's start with a basic mathematical function.

The syntax is as follows:

    def function(a, b, c):
        code here
        return
        
def defines the function, function is the name of the function that you'll use when you call it, and (a, b, c) are the arguments required for the function to run. "Return" is what you want the code to spit out - in the case of the mathematical function, it's the result of the expression.
 

```python
def y(m, b, x):
    quad = m * x**2
    return quad + b

#Can also write it like this:

def y_alt(m, b, x):
    quad = m * x**2
    function = quad + b
    return function
```

This returns a simple quadratic curve, given a slope and a y-axis intersection, as well as a sample of x's. So to run this, we might give it a single value of m and b, and an array for x.

```python
xarray = np.linspace(0, 20, 21)

line = y(2, 0, xarray)
print(line)

#line2 = y_alt(2, 0, xarray)
```

# Gaussian Statistics

## Histograms

In astronomy, one of the most frequent tasks you'll do is bin your data into a histogram. Luckily, Python has a lot of flexibility to perform basic statistics.

Of course you can always plot real data points, but for now let's generate some Gaussian distributions to illustrate our point.

There are a couple of ways to generate a Gaussian: randn and normal. Randn samples from the standard Normal - where $\mu$=0 and $\sigma$=1. Its argument is the sample size. Normal allows you to change mu and sigma, so the arguments are mu, sigma, sample size.

```python
gaussian = np.random.normal(0, 2, 10000) #Gaussian distribution centered on 0 
                                            #with a std. dev. of 2

plt.hist(gaussian, bins=50, normed=True)
plt.grid(True)
plt.show()
```

While we're at it, let's go over some cool plotting tips:

    facecolor: takes a string argument specifying the color of the histogram
    alpha: goes from 0 to 1, specifies the opacity. helpful when overlaying multiple histograms
    hatch: arguments such as '//', '\\', '.', '..' will fill in the histogram with this style
    histtype: can be things like bar, step. specifies the style of the histogram
    normed: boolean; True means normalize to 1
    label: a string argument labeling the subplot. to activate the legend, type plt.legend() 
    

Now let's actually fit a Gaussian curve to these distributions. Since we randomly generated these, one would hope that it fits it to the $\mu$ and $\sigma$ values we gave it.

```python
#First let's fit the data.
from scipy.stats import norm
norm
```

This is a new function that we haven't used before. If you're ever unsure about how to use a function, you can type ? after the function name to call up the docstrings.

```python
mu, sigma = norm.fit(gaussian)
print(mu, sigma)

#Now let's calculate the PDF and plot it.
x = np.linspace(-6, 6, 10000)
p = norm.pdf(x, mu, sigma)
plt.plot(x, p, 'k', linewidth=2)


plt.hist(gaussian, bins=50, color='m', alpha=0.4, normed=True, fill=True, histtype='step')
plt.title("Fit results: $\mu$ = %.2f,  $\sigma$ = %.2f" % (mu, sigma))
plt.show()
```

```python
def gauss(x, f0, mu, sig):
    const = f0/(sig**2 * np.sqrt(2*np.pi))
    exp = np.exp(-(x-mu)**2/(2*sig**2))
    return const*exp
```

The complicated part is learning how to use fits in Python, but just follow this syntax. It fits the data, and saves the best fit parameters as well as the covariance matrix, but don't worry too much about that. All we want to do is fit a Gaussian to our emission line and have it spit out a reasonable error.

```python
bfpars, covar = opt.curve_fit(gauss, wavelength, flux, p0=[1.7e-18,10700.,50.], sigma=fluxerr)
#p0 is an initial guess: flux, wavelength, width.
print(bfpars)
```
